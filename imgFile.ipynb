```python
import matplotlib.pyplot as plt
import sys
```

```python
def display_image_inline(path: str):
    """
    Read and display an image inline using matplotlib.
    """
    img = plt.imread(path)
    plt.imshow(img)
    plt.axis('off')  # hide axes
    plt.show()

display_image_inline("abo-images-small/images/small/8c/8ccb5859.jpg")
```
